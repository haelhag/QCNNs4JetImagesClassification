```python
from math import sqrt
from numpy import dot

p1 = [First_ten_part[0,1],First_ten_part[0,2],First_ten_part[0,3]]
print('p1 =',p1)
p2 = [First_ten_part[1,1],First_ten_part[1,2],First_ten_part[1,3]]
print('p1 =',p2)
p3 = [First_ten_part[2,1],First_ten_part[2,2],First_ten_part[2,3]]
print('p3 =',p3)

PJ = [p1[0] + p2[0] + p3[0], p1[1] + p2[1] + p3[1], p1[2] + p2[2] + p3[2]]

print('PJ =', PJ)
```

```python
#Ei:
p01 = [First_ten_part[0,0]]
print('p01 =',p01)
p02 = [First_ten_part[1,0]]
print('p02 =',p02)
p03 = [First_ten_part[2,0]]
print('p03 =',p03)

#EJ:
P0J = [p01[0] + p02[0] + p03[0]]
print('P0J =', P0J)

Mag_PJ = sqrt(PJ[0]*PJ[0]+PJ[1]*PJ[1]+PJ[2]*PJ[2])
print('|PJ| =',Mag_PJ)
```

```python
#construction of GS basis:
ep1 = [PJ[0]/Mag_PJ,PJ[1]/Mag_PJ,PJ[2]/Mag_PJ]
print('ep1 =', ep1)

ep1_dot_p1 = dot(ep1,p1)
print('ep1_dot_p1 =', ep1_dot_p1)

ep2_Nume = [p1[0]-(ep1_dot_p1*ep1[0]), p1[1]-(ep1_dot_p1*ep1[1]), p1[2]-(ep1_dot_p1*ep1[2])]
print('ep2_Nume =', ep2_Nume)

ep2_Deno = sqrt(dot(ep2_Nume,ep2_Nume))
print('ep2_Deno =', ep2_Deno)

ep2 = [ep2_Nume[0]/ep2_Deno, ep2_Nume[1]/ep2_Deno, ep2_Nume[2]/ep2_Deno]
print('ep2 =',ep2)

ep1_dot_p2 = dot(ep1,p2)

ep2_dot_p2 = dot(ep2,p2)

ep3_Nume = [ p2[0]-(ep1_dot_p2*ep1[0])-(ep2_dot_p2*ep2[0]), p2[1]-(ep1_dot_p2*ep1[1])-(ep2_dot_p2*ep2[1]), p2[2]-(ep1_dot_p2*ep1[2])-(ep2_dot_p2*ep2[2])]
ep3_Deno = sqrt(dot(ep3_Nume,ep3_Nume))

ep3 = [ep3_Nume[0]/ep3_Deno, ep3_Nume[1]/ep3_Deno, ep3_Nume[2]/ep3_Deno]
print('ep3 =',ep3)

print('dot(ep1,ep2)=',dot(ep1,ep2))
print('dot(ep1,ep3)=',dot(ep1,ep3))
print('dot(ep2,ep3)=',dot(ep2,ep3))
```

```python
import matplotlib.pyplot as plt
#Lorentz boost:

Njets = 10000
l = 0
gamm = 2
for k in range(10):
    X_avg_top = []
    Y_avg_top = []
    W_avg_top = []
    X_avg_qcd = []
    Y_avg_qcd = []
    W_avg_qcd = []
    print(k)

    for j in range(l,l+Njets):
        #print(j)
        First_ten_part = particle_data[j,: 200]
        #print('jet type:', jet_data[j,0])
    
        p1 = [First_ten_part[0,1],First_ten_part[0,2],First_ten_part[0,3]]
        p2 = [First_ten_part[1,1],First_ten_part[1,2],First_ten_part[1,3]]
        p3 = [First_ten_part[2,1],First_ten_part[2,2],First_ten_part[2,3]]
        PJ = [p1[0] + p2[0] + p3[0], p1[1] + p2[1] + p3[1], p1[2] + p2[2] + p3[2]]
        
        for m in range(3, len(First_ten_part)):
            PJ[0] = PJ[0]+ First_ten_part[m,1]
            PJ[1] = PJ[1]+ First_ten_part[m,2]
            PJ[2] = PJ[2]+ First_ten_part[m,3]
    
        #Ei:
        p01 = [First_ten_part[0,0]]
        p02 = [First_ten_part[1,0]]
        p03 = [First_ten_part[2,0]]

        #EJ:
        P0J = [p01[0] + p02[0] + p03[0]]
        Mag_PJ = sqrt(PJ[0]*PJ[0]+PJ[1]*PJ[1]+PJ[2]*PJ[2])
        Mag_P0J = sqrt(dot(P0J,P0J))
    
        m0 = Mag_PJ/gamm
    
        P4J = [p01[0] + p02[0] + p03[0], -(p1[0] + p2[0] + p3[0]),-(p1[1] + p2[1] + p3[1]), -(p1[2] + p2[2] + p3[2])]
        #print(P4J)
    
        mJ = sqrt(dot(P4J,P4J))
        #print(mJ)
    
        p1_B = [m0/mJ * First_ten_part[0,1],m0/mJ * First_ten_part[0,2],m0/mJ * First_ten_part[0,3]]
        p2_B = [m0/mJ * First_ten_part[1,1],m0/mJ * First_ten_part[1,2],m0/mJ * First_ten_part[1,3]]
        p3_B = [m0/mJ * First_ten_part[2,1],m0/mJ * First_ten_part[2,2],m0/mJ * First_ten_part[2,3]]
        p01_B = [m0/mJ * First_ten_part[0,0]]
        p02_B = [m0/mJ * First_ten_part[1,0]]
        p03_B = [m0/mJ * First_ten_part[2,0]]
        #PJ_B = [m0/mJ * (p1[0] + p2[0] + p3[0]), m0/mJ * (p1[1] + p2[1] + p3[1]), m0/mJ * (p1[2] + p2[2] + p3[2])]
        PJ_B = [m0/mJ * PJ[0], m0/mJ * PJ[1], m0/mJ * PJ[2]]
        P0J_B = [m0/mJ * (p01[0] + p02[0] + p03[0])]
        Mag_PJ_B = m0/mJ * Mag_PJ

        #construction of GS basis:
        ep1 = [PJ_B[0]/Mag_PJ_B,PJ_B[1]/Mag_PJ_B,PJ_B[2]/Mag_PJ_B]
        ep1_dot_p1 = dot(ep1,p1_B)

        ep2_Nume = [p1_B[0]-(ep1_dot_p1*ep1[0]), p1_B[1]-(ep1_dot_p1*ep1[1]), p1_B[2]-(ep1_dot_p1*ep1[2])]
        ep2_Deno = sqrt(dot(ep2_Nume,ep2_Nume))
        ep2 = [ep2_Nume[0]/ep2_Deno, ep2_Nume[1]/ep2_Deno, ep2_Nume[2]/ep2_Deno]

        ep1_dot_p2 = dot(ep1,p2_B)
        ep2_dot_p2 = dot(ep2,p2_B)
        ep3_Nume = [ p2_B[0]-(ep1_dot_p2*ep1[0])-(ep2_dot_p2*ep2[0]), p2_B[1]-(ep1_dot_p2*ep1[1])-(ep2_dot_p2*ep2[1]), p2_B[2]-(ep1_dot_p2*ep1[2])-(ep2_dot_p2*ep2[2])]
        ep3_Deno = sqrt(dot(ep3_Nume,ep3_Nume))

        ep3 = [ep3_Nume[0]/ep3_Deno, ep3_Nume[1]/ep3_Deno, ep3_Nume[2]/ep3_Deno]
    
        #Jet images 2D coordinates:
        X = []
        Y = []
        W = []
    
        for i in range(len(First_ten_part)):
            X.append(dot([m0/mJ * First_ten_part[i,1],m0/mJ * First_ten_part[i,2],m0/mJ * First_ten_part[i,3]],ep2)/(m0/mJ * First_ten_part[0,0]))
            Y.append(dot([m0/mJ * First_ten_part[i,1],m0/mJ * First_ten_part[i,2],m0/mJ * First_ten_part[i,3]],ep3)/(m0/mJ * First_ten_part[0,0]))
            W.append(m0/mJ * First_ten_part[i,0]/(p01_B[0] + p02_B[0] + p03_B[0]))
            #W.append(First_ten_part[i,0]/(p01_B[0] + p02_B[0] + p03_B[0]))
            if jet_data[j,0] == 0.0:
                #print(jet_data[j])
                X_avg_qcd.append((dot([m0/mJ * First_ten_part[i,1],m0/mJ * First_ten_part[i,2],m0/mJ * First_ten_part[i,3]],ep2)/(m0/mJ * First_ten_part[0,0])))#/Njets )
                Y_avg_qcd.append((dot([m0/mJ * First_ten_part[i,1],m0/mJ * First_ten_part[i,2],m0/mJ * First_ten_part[i,3]],ep3)/(m0/mJ * First_ten_part[0,0])))#/Njets )
                W_avg_qcd.append( (m0/mJ * First_ten_part[i,0]/(p01_B[0] + p02_B[0] + p03_B[0]))/len(First_ten_part))#/Njets )
                #W_avg_qcd.append(First_ten_part[i,0]/(p01_B[0] + p02_B[0] + p03_B[0]))
            # elif jet_data[j,0] == 1.0:
            #     #print(jet_data[j])
            #     X_avg_top.append((dot([m0/mJ * First_ten_part[i,1],m0/mJ * First_ten_part[i,2],m0/mJ * First_ten_part[i,3]],ep2)/(m0/mJ * First_ten_part[0,0])))#/Njets )
            #     Y_avg_top.append((dot([m0/mJ * First_ten_part[i,1],m0/mJ * First_ten_part[i,2],m0/mJ * First_ten_part[i,3]],ep3)/(m0/mJ * First_ten_part[0,0])))#/Njets )
            #     W_avg_top.append(((m0/mJ * First_ten_part[i,0]/(p01_B[0] + p02_B[0] + p03_B[0]))/len(First_ten_part)))#/Njets )
            #     #W_avg_top.append(First_ten_part[i,0]/(p01_B[0] + p02_B[0] + p03_B[0]))
    l=l+Njets
                

    #print('X =',X)
    #print('Y =',Y)
    #print('W =',W)
    if len(X_avg_qcd) != 0:
        plt.hist2d(X_avg_qcd, Y_avg_qcd, bins=(28, 28), weights=W_avg_qcd, range=[[-0.02, 0.02], [-0.02, 0.02]])#, normed=True)#, cmap=plt.cm.jet, density=True)
        #plt.savefig(f"/Users/halaelhag/mnt/droplet/topQcdJets_Avg_Wbkg/qcd/jetqcd{k+1}.png")
        plt.colorbar()
        plt.savefig(f"/Users/halaelhag/mnt/droplet/topQcdJets_Avg_bars_Modi/qcd/jetqcd{k+1}.png")
        plt.show()
        plt.close()
    
    # if len(X_avg_top) != 0:
    #     plt.hist2d(X_avg_top, Y_avg_top, bins=(20, 20), weights=W_avg_top, range=[[-0.02, 0.02], [-0.02, 0.02]])#, normed=True)#, cmap=plt.cm.jet, density=True)
    #     #plt.savefig(f"/Users/halaelhag/mnt/droplet/topQcdJets_Avg_Wbkg/top/jettop{k+1}.png")
    #     plt.colorbar()
    #     plt.savefig(f"/Users/halaelhag/mnt/droplet/topQcdJets_Avg_bars_Modi/top/jettop{k+1}.png")
    #     plt.show()
    #     plt.close()
```
